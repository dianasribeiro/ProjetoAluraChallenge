### Importação dos dados



```python
# Carregando os arquivos csv das lojas

import pandas as pd

loja1 = pd.read_csv('loja_1.csv')
loja2 = pd.read_csv('loja_2.csv')
loja3 = pd.read_csv('loja_3.csv')
loja4 = pd.read_csv('loja_4.csv')

```

```python
display(loja2)
```

#1. Análise do faturamento


```python
# Analisando o faturamento de cada loja

faturamento_loja1 = loja1['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

print(f' Faturamento total loja 1: {faturamento_loja1}')
print(f' Faturamento total loja 2: {faturamento_loja2}')
print(f' Faturamento total loja 3: {faturamento_loja3}')
print(f' Faturamento total loja 4: {faturamento_loja4}')
```

```python
# Faturamento total de todas as lojas

faturamento_total = faturamento_loja1 + faturamento_loja2 + faturamento_loja3 + faturamento_loja4

print(f' Faturamento total de todas as lojas: {faturamento_total}')
```

# 2. Vendas por Categoria


```python
vendas_por_categoria_loja1 = loja1.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja2 = loja2.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja3 = loja3.groupby('Categoria do Produto')['Preço'].sum()
vendas_por_categoria_loja4 = loja4.groupby('Categoria do Produto')['Preço'].sum()

print(f'As vendas por categoria da loja 1 são: \n{vendas_por_categoria_loja1}\n')
print(f'As vendas por categoria da loja 2 são: \n{vendas_por_categoria_loja2}\n')
print(f'As vendas por categoria da loja 3 são: \n{vendas_por_categoria_loja3}\n')
print(f'As vendas por categoria da loja 4 são: \n{vendas_por_categoria_loja4}\n')
```

```python
# Venda total por categoria de todas as lojas

# Concatenando os DataFrames
todas_as_lojas = pd.concat([loja1, loja2, loja3, loja4])

#Agrupando por categoria e somando as vendas
vendas_por_categoria = todas_as_lojas.groupby('Categoria do Produto')['Preço'].sum()

print(vendas_por_categoria)
```

```python

# Encontrando a categoria com maior valor de venda e mais popular
categoria_mais_vendida = vendas_por_categoria.idxmax()


# Encontrando a categoria com menor valor de venda e menos popular
categoria_menos_vendida = vendas_por_categoria.idxmin()


print(f'A categoria mais popular entre todas as lojas é: {categoria_mais_vendida}\n')

print(f'A categoria menos popular entre todas as lojas é: {categoria_menos_vendida}')



```

# 3. Média de Avaliação das Lojas

```python
# Calculando a media de avaliação dos clientes para cada loja
media_avaliacao_loja1 = loja1['Avaliação da compra'].mean()
media_avaliacao_loja2 = loja2['Avaliação da compra'].mean()
media_avaliacao_loja3 = loja3['Avaliação da compra'].mean()
media_avaliacao_loja4 = loja4['Avaliação da compra'].mean()

print(f' A media de avaliação dos clientes da loja 1 é: {media_avaliacao_loja1:.3f}')
print(f' A media de avaliação dos clientes da loja 2 é: {media_avaliacao_loja2:.3f}')
print(f' A media de avaliação dos clientes da loja 3 é: {media_avaliacao_loja3:.3f}')
print(f' A media de avaliação dos clientes da loja 4 é: {media_avaliacao_loja4:.3f}')
```

```python
# calculando a media de avaliação dos clientes entre todas as lojas
media_avaliacao_lojas = todas_as_lojas['Avaliação da compra'].mean()
print(f' A media de avaliação dos clientes entre todas as lojas é: {media_avaliacao_lojas:.3f}')
```

# 4. Produtos Mais e Menos Vendidos

```python
# Encontrando a categoria mais vendida
valor_mais_vendido = vendas_por_categoria.max()
valor_menos_vendido = vendas_por_categoria.min()

print(f'O valor de vendas da categoria mais vendida é: {valor_mais_vendido}\n')
print(f'O valor de vendas da categoria menos vendida é: {valor_menos_vendido}')

```

# 5. Frete Médio por Loja


```python
# Calculando o custo medio do frete
custo_frete_loja1 = loja1['Frete'].mean()
custo_frete_loja2 = loja2['Frete'].mean()
custo_frete_loja3 = loja3['Frete'].mean()
custo_frete_loja4 = loja4['Frete'].mean()

print(f' O custo medio do frete da loja 1 é {custo_frete_loja1:.2f}')
print(f' O custo medio do frete da loja 2 é {custo_frete_loja2:.2f}')
print(f' O custo medio do frete da loja 3 é {custo_frete_loja3:.2f}')
print(f' O custo medio do frete da loja 4 é {custo_frete_loja4:.2f}')
```

# Avaliação dos clientes por loja

```python
# Fazendo o grafico pie para saber a media de avaliação por loja dos clientes

import matplotlib.pyplot as plt

# Dados para o gráfico
labels = ['loja1', 'loja2', 'loja3', 'loja4']
sizes = [media_avaliacao_loja1,media_avaliacao_loja2,media_avaliacao_loja3,media_avaliacao_loja4]  # Valores correspondentes a cada fatia

#Definindo as cores com plot_color_gradients
plot_color_gradients = ('Sequential',
                     ['Greys', 'Purples', 'Blues', 'Greens', 'Oranges', 'Reds',
                      'YlOrBr', 'YlOrRd', 'OrRd', 'PuRd', 'RdPu', 'BuPu',
                      'GnBu', 'PuBu', 'YlGnBu', 'PuBuGn', 'BuGn', 'YlGn'])  # Cores das fatias

# Usando cmap para aplicar as cores
cmap = plt.get_cmap(plot_color_gradients[1][7]) # usandso a cor YlOrRd
colors = [cmap(i / len(labels) * 0.5) for i in range(len(labels))]
explode = (0, 0, 0.1, 0)  # Destacando uma fatia (loja3) que tem a melhor avaliação

# Criando o gráfico de pizza
plt.pie(sizes, explode=explode, labels=labels,colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)

# Adicionando um título
plt.title('Avaliação dos Clientes por Loja')

# Mostrando o gráfico
plt.show()
```
